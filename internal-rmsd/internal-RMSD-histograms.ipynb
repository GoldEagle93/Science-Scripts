```python
import os
import numpy as np
import pandas as pd
from os.path import exists
import matplotlib.pyplot as plt
```

```python
source = '/Users/HQ/orange/Reza/TF/NAR/' #14-two-sided-dummy-contacts/'
systems = sorted([i for i in os.listdir(source) if '.' not in i])
systems = ['1a74', '1azp', '1by4', '1bgb',  '2b0d', '1cdw', '1dh3', '1jj4', '1r4o', '1r4r', '1zme', '3cro', '2r1j', '2dgc', '1ysa']
# systems = ['1a74', '1azp', '1by4',  '2b0d', '1cdw', '1dh3', '1jj4', '1r4o', '1r4r', '1zme', '2dgc', '2r1j', '3cro', '1ysa']
# systems = ['1bgb',  '2b0d']
bold = ['#fcff5d', '#7dfc00', '#0ec434', '#228c68', '#8ad8e8', '#235b54', '#29bdab', '#3998f5', '#37294f', '#277da7', '#3750db', '#f22020', '#991919', '#ffcba5', '#e68f66', '#c56133', '#96341c', '#632819', '#ffc413', '#f47a22', '#2f2aa0', '#b732cc', '#772b9d', '#f07cab', '#d30b94', '#edeff3', '#c3a5b4', '#946aa2', '#5d4c86', '#201923']
uf = ['#0021A5', '#F2A900', '#FA4616', '#22884C']
os.chdir(source)
```

```python
for i in range(len(systems)):
    fig, axes = plt.subplots(nrows=6,ncols=5,sharex=True,sharey=True,figsize=(20,8))
    fig.subplots_adjust(wspace=0, hspace=0)
    if exists('%s/14-two-sided-dummy-contacts/rmsd-DNA.dat'%(systems[i])):
        df = pd.read_csv('%s/14-two-sided-dummy-contacts/rmsd-DNA.dat'%(systems[i]), delim_whitespace=True)
        for j in range(len(df.columns)-1):
            row = int(j/5)
            column = j%5
            axes[row,column].hist(df[df.columns[j+1]], bins=120, range=(0, 30), color=uf[0])
            if column == 0:  axes[row,column].set_ylabel('Population')
            if row == 5:     axes[row,column].set_xlabel('Internal DNA RMSD')
    # plt.savefig('RMSD-pd-histogram-%s.png'%(systems[i]), bbox_inches='tight', dpi=300)
    plt.show()
    plt.clf()
```

```python
#DNA
fig, axes = plt.subplots(nrows=3,ncols=5,sharex=True,sharey=True,figsize=(20,8))
# fig.subplots_adjust(wspace=0, hspace=0)
for i in range(len(systems)):
    row = int(i/5)
    column = i%5
    if exists('%s/14-two-sided-dummy-contacts/rmsd-DNA.dat'%(systems[i])):
        df = pd.read_csv('%s/14-two-sided-dummy-contacts/rmsd-DNA.dat'%(systems[i]), delim_whitespace=True)
        for j in range(len(df.columns)-1):
            axes[row,column].hist(df[df.columns[j+1]], bins=120, range=(0, 15), color=uf[1])
        axes[row,column].hist(df[df.columns[1]], bins=120, range=(0, 15), color=uf[0])
        axes[row,column].set_title(systems[i])
    if column == 0:  axes[row,column].set_ylabel('Population')
    if row == 2:     axes[row,column].set_xlabel('Internal DNA RMSD (Ã…)')
    # plt.clf()
plt.savefig('internal-DNA-RMSD.png', bbox_inches='tight', dpi=300)
# plt.show()
```
