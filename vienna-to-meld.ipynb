```python
sequence =   'GGGAGUUAGACUCGCGAAAGCGAGGAACUGCGAAAGCAGUACUUCC'
ss =         '((((((....(((((....)))))..(((((....)))))))))))' #RNAFold
#rnaComposer '((((((....(((((....)))))..(((((....)))))))))))'

f = list(range(len(ss)))
b = f.copy()
b.reverse()

if len(sequence) != len(ss):
    print('length mismatch')

forwards = []
backwards = []
periods = []
for i in f:
    if ss[i] == '(':
        forwards.append(i)
    if ss[i] == ')':
        backwards.append(i)
    if ss[i] == '.':
        periods.append(i)
forwards.reverse()

pairs = [[], []]
for i in forwards[:]:
    pairs[0].append(i)
    for j in backwards:
        if j > i and j not in pairs[1]:
            pairs[1].append(j)
            break

ones = pairs[0].copy()
twos = pairs[1].copy()
ones = [i+1 for i in ones]
twos = [i+1 for i in twos]
ones.reverse()
twos.reverse()

for i in range(len(ones)):
    if sequence[ones[i]-1] == 'A' and sequence[twos[i]-1] == 'U':
        print(
'''%s   N6    %s    O4     2.9
%s   N3    %s    N1     2.9
%s   C4'   %s    C4'    15
              '''%(ones[i], twos[i], ones[i], twos[i], ones[i], twos[i]))
    elif sequence[ones[i]-1] == 'U' and sequence[twos[i]-1] == 'A':
        print(
'''%s   O4    %s    N6     2.9
%s   N1    %s    N3     2.9
%s   C4'   %s    C4'    15
              '''%(ones[i], twos[i], ones[i], twos[i], ones[i], twos[i]))
    elif sequence[ones[i]-1] == 'C' and sequence[twos[i]-1] == 'G':
        print(
'''%s   N3    %s    N1     2.9
%s   O2    %s    N2     2.9
%s   N4    %s    O6     2.9
%s   C4'   %s    C4'    15
              '''%(ones[i], twos[i], ones[i], twos[i], ones[i], twos[i], ones[i], twos[i]))
    elif sequence[ones[i]-1] == 'G' and sequence[twos[i]-1] == 'C':
        print(
'''%s   N1    %s    N3     2.9
%s   N2    %s    O2     2.9
%s   O6    %s    N4     2.9
%s   C4'   %s    C4'    15
              '''%(ones[i], twos[i], ones[i], twos[i], ones[i], twos[i], ones[i], twos[i]))
```
